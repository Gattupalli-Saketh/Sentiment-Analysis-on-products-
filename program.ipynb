```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from transformers import pipeline

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load your DataFrame (replace 'filtered_reviews.csv' with your actual file)
df = pd.read_csv('filtered_reviews.csv')

# Display the first 100 rows of the DataFrame
example = df.head(100)
print(example)

# Preprocessing function
def preprocess_text(text):
    # Tokenize the input text
    tokens = word_tokenize(text)  # Tokenize the text
    
    # Get the list of English stopwords
    stop_words = set(stopwords.words('english'))
    
    # Remove stopwords from tokens
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    
    return ' '.join(filtered_tokens)  # Join tokens back into a string

# Ensure the column name matches your dataset (e.g., 'ReviewText' or 'Review Text')
if 'Review Text' in df.columns:
    review_column = 'Review Text'
elif 'ReviewText' in df.columns:
    review_column = 'ReviewText'
else:
    raise ValueError("Review column not found in the dataset. Please check the column name.")

# Apply preprocessing to the review column
df['cleaned_token'] = df[review_column].dropna().apply(preprocess_text)

# Display the first few rows of the DataFrame with cleaned reviews
print(df[[review_column, 'cleaned_token']].head())

# Sentiment Analysis
sentiment_analysis = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english") 

# Add sentiment labels to the DataFrame
def get_sentiment(text):
    if text.strip():  # Check if text is not empty
        return sentiment_analysis(text)[0]['label']
    else:
        return "NEUTRAL"  # Default label for empty or invalid text

df['sentiment'] = df['cleaned_token'].apply(get_sentiment)

# Display the first few rows with sentiment labels
print(df[['cleaned_token', 'sentiment']].head())
```
